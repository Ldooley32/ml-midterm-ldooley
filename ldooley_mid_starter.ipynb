```python
import pandas as pd

df = pd.read_csv("train.csv")

df.head(n=30)

```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
sex = df['Sex']
survived = df['Survived']

# Create DataFrames for counts
survival_counts = df[df['Survived'] == 1]['Sex'].value_counts()
non_survival_counts = df[df['Survived'] == 0]['Sex'].value_counts()

# Combine the counts into a DataFrame
counts_df = pd.DataFrame({'Survived': survival_counts, 'Not Survived': non_survival_counts}).fillna(0)

# Plot the bar plot
counts_df.plot(kind='bar')

plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survivors and Non-Survivors by Sex')
plt.show()
```

#### Fix missing values for age
Replace missing values for age with the average (mean).  Do the operation in place.

```python
# Calculate the mean age
mean_age = df['Age'].mean()

# Replace missing values in the 'Age' column with the mean age
df['Age'] = df['Age'].fillna(mean_age)
```

```python
# check that cleaning is good (You don't need to add anything here, but you may)
df.head(n=30)
```

### Decision Tree classifier

I chose the features for you!

```python

from sklearn.model_selection import StratifiedShuffleSplit

# Create the feature matrix and target vector
X = df[['Age', 'Fare']]
y = df['Survived']

# Initialize the StratifiedShuffleSplit object
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=123)

# Perform the split
for train_indices, test_indices in splitter.split(X, y):
    train_set = df.iloc[train_indices]
    test_set = df.iloc[test_indices]
# Check the results
print("Training set size:", len(train_set))
print("Test set size:", len(test_set))



```

Build and train a decision tree classifier
Print the confusion matrix and scores for the trained model

Use the variable y_predicted to determine the performance of the model.  The graphing code below needs to have it defined in order to work correctly.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

# Assuming train_set and test_set are already defined

X_train = train_set[['Age', 'Fare']]
y_train = train_set['Survived']

X_test = test_set[['Age', 'Fare']]  # Corrected indexing for X_test
y_test = test_set['Survived']  # Corrected typo in 'Survived'

tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)




y_pred = tree_model.predict(X_train)

print('Results for decision tree on training data')
print('Default settings')
print("Confusion Matrix")
print(confusion_matrix(y_train, y_pred))
print('Accuracy is ', accuracy_score(y_train, y_pred))
print('Precision is ', precision_score(y_train, y_pred))
print('Recall is ', recall_score(y_train, y_pred))
print('F1 is ', f1_score(y_train, y_pred))
```

```python
y_test_pred = tree_model.predict(X_test)

print('Results for decision tree on test data')
print('  Default settings')
print("Confusion Matrix ")
print(confusion_matrix(y_test, y_test_pred))
print('Accuracy is  ', accuracy_score(y_test, y_test_pred))
print('Precision is  ', precision_score(y_test, y_test_pred))
print('Recall is  ', recall_score(y_test, y_test_pred))
print('F1 is ', f1_score(y_test, y_test_pred))
```
